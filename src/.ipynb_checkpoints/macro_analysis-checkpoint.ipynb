This report investigates cross-asset relationships between Gold, the US Dollar Index (DXY), and US 10Y Treasury yields using time-series analysis from 2020 to 2026.

The objective is to evaluate:

- Correlation dynamics  
- Volatility regimes  
- Risk-adjusted performance  
- Drawdown characteristics  
- Macro-hedging properties  

# Executive Summary

Key findings:

• Gold exhibits persistent negative correlation with DXY (~ -0.38).  
• 60-day rolling correlation strengthened to approximately -0.45 in early 2026.  
• Gold displays higher short-term volatility (~3%) compared to DXY (~0.4%).  
• Gold delivered superior risk-adjusted returns (Sharpe ≈ 1.19).  
• US10Y experienced significantly deeper drawdowns (~ -73%).  

These results confirm Gold’s macro-sensitive nature and partial hedge characteristics against USD strength.

# Executive Summary

Key findings:

• Gold exhibits persistent negative correlation with DXY (~ -0.38).  
• 60-day rolling correlation strengthened to approximately -0.45 in early 2026.  
• Gold displays higher short-term volatility (~3%) compared to DXY (~0.4%).  
• Gold delivered superior risk-adjusted returns (Sharpe ≈ 1.19).  
• US10Y experienced significantly deeper drawdowns (~ -73%).  

These results confirm Gold’s macro-sensitive nature and partial hedge characteristics against USD strength.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

prices = pd.read_csv("../data/prices.csv", index_col=0, parse_dates=True)
returns = pd.read_csv("../data/returns.csv", index_col=0, parse_dates=True)

prices.head()
```

```python
correlation = returns.corr()
correlation
```

```python
rolling_corr_60 = returns["GOLD"].rolling(60).corr(returns["DXY"])
rolling_vol_30 = returns.rolling(30).std()

# Rolling correlation plot
plt.figure(figsize=(10, 5))
plt.plot(rolling_corr_60)
plt.title("60-Day Rolling Correlation: GOLD vs DXY")
plt.show()

# Rolling volatility plot
plt.figure(figsize=(10, 5))
plt.plot(rolling_vol_30["GOLD"], label="GOLD")
plt.plot(rolling_vol_30["DXY"], label="DXY")
plt.plot(rolling_vol_30["US10Y"], label="US10Y")
plt.legend()
plt.title("30-Day Rolling Volatility")
plt.show()
```

```python
annual_return = returns.mean() * 252
annual_vol = returns.std() * np.sqrt(252)
sharpe = annual_return / annual_vol

cumulative = (1 + returns).cumprod()
drawdown = cumulative / cumulative.cummax() - 1
max_drawdown = drawdown.min()

summary = pd.DataFrame({
    "AnnualReturn": annual_return,
    "AnnualVol": annual_vol,
    "Sharpe": sharpe,
    "MaxDrawdown": max_drawdown
})

summary
```

# Interpretation

Gold maintains a structurally negative relationship with the US Dollar Index.

During periods of intensified USD strength, gold prices tend to face downward pressure.

Rolling analysis reveals regime-dependent variation in correlation intensity, particularly visible in early 2026.

Despite higher volatility, gold delivered superior risk-adjusted returns, suggesting compensation for elevated risk exposure.

US10Y displayed significant drawdown risk, highlighting duration sensitivity during stress environments.

```python
median_vol = rolling_vol_30["GOLD"].median()

high_vol = rolling_vol_30["GOLD"] > median_vol
low_vol = rolling_vol_30["GOLD"] <= median_vol

high_vol_return = returns["GOLD"][high_vol].mean()
low_vol_return = returns["GOLD"][low_vol].mean()

high_vol_return, low_vol_return
```

# Volatility Regime Insight

Gold demonstrates different return behavior across volatility regimes.

High-volatility periods often coincide with macro stress environments, reinforcing gold’s defensive characteristics.

# Conclusion

This cross-asset analysis confirms gold’s macro-driven sensitivity and partial hedge properties.

While exhibiting higher short-term volatility, gold delivered stronger risk-adjusted performance relative to DXY and US10Y.

These findings reinforce gold’s relevance in diversified macro portfolios.
